```python
import numpy as np
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/CV_images/avengerexam.jpeg')
```

```python
cv2_imshow(image)
```

```python
print('Image Width is',image.shape[1])
print('Image Height is',image.shape[0])
```

```python
img_negative = 255 - image
```

```python
plt.imshow(img_negative, cmap='gray')
plt.title('Image Negative')
```

```python
img_8bit = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hist_eq = cv2.equalizeHist(img_8bit)
```

```python
plt.hist(image.flatten(), bins=256, range=[0, 256], color='black', alpha=0.7)
plt.title('Original Image Histogram')
```

```python
plt.imshow(hist_eq, cmap='gray')
plt.title('Histogram Equalization')
```

```python
plt.hist(hist_eq.flatten(), bins=256, range=[0, 256], color='black', alpha=0.7)
plt.title('Equalized Image Histogram')
```

```python
lt = 100
ut = 200
gls = np.copy(image)
gls[(image >= lt) & (image <= ut)] = 255
gls[(image < lt) | (image > ut)] = 0
```

```python
plt.imshow(gls, cmap='gray')
plt.title('Gray Level Slicing')
```

```python
gbk = np.array([[1, 4, 6, 4, 1],
                             [4, 16, 24, 16, 4],
                             [6, 24, 36, 24, 6],
                             [4, 16, 24, 16, 4],
                             [1, 4, 6, 4, 1]])/256
```

```python
img = cv2.resize(image, (500,300))
```

```python
gbimage = cv2.filter2D(src=img, ddepth=-1, kernel=gbk)
```

```python
cv2_imshow(gbimage)
```

```python
imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
lap = cv2.Laplacian(imggray, cv2.CV_64F)
```

```python
lap = np.uint8(np.absolute(lap))
```

```python
cv2_imshow(lap)
```
