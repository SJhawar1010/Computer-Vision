```python
image = cv2.imread('/content/drive/MyDrive/CV_images/avg2.jpg')
```

```python
def linear_filter(image, kernel):
    height, width = image.shape[:2]
    k_height, k_width = kernel.shape[:2]
    pad_height = k_height // 2
    pad_width = k_width // 2
    padded_image = cv2.copyMakeBorder(image, pad_height, pad_height, pad_width, pad_width, cv2.BORDER_REPLICATE)
    output_image = np.zeros_like(image, dtype=np.float32)
    for y in range(height):
        for x in range(width):
            for c in range(image.shape[2]):
                output_image[y, x, c] = np.sum(padded_image[y:y+k_height, x:x+k_width, c] * kernel)

    return np.clip(output_image, 0, 255).astype(np.uint8)
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
```

```python
filtered_image = linear_filter(image, kernel)
```

```python
cv2_imshow(image)
```

```python
cv2_imshow(filtered_image)
```
