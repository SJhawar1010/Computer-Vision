```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/content/drive/MyDrive/CV_images/avengerexam.jpeg',0)
```

```python
img
```

```python
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
```

```python
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(np.abs(sobel_x), cmap='gray')
plt.title('Sobel X')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(np.abs(sobel_y), cmap='gray')
plt.title('Sobel Y')
plt.show()
```

Sobel X and Sobel Y images show the gradients of the original image along the horizontal and vertical directions, respectively.

Edges are detected where the gradient magnitude is high in either the horizontal or vertical direction.

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
```

```python
plt.imshow(np.abs(laplacian), cmap='gray')
plt.title('Laplacian')
plt.show()
```

The Laplacian image highlights regions of rapid intensity change, typically where the second derivative of intensity with respect to position reaches a maximum or minimum.

Edges are detected where the Laplacian values are large.

```python
edges = cv2.Canny(img, 100, 200)
```

```python
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.show()
```

The Canny edge detector is an optimal edge detector that combines first-order derivatives (Sobel operator) with hysteresis thresholding to accurately detect edges.

It produces thin edges that accurately localize the boundaries of objects in the image.
