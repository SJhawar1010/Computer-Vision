```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python
train_path = '/content/drive/MyDrive/CV_images/Images/Train'
test_path = '/content/drive/MyDrive/CV_images/Images/Test'
```

```python
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    resized_image = cv2.resize(image, (100, 100))
    grayscale_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
    hist = cv2.calcHist([grayscale_image], [0], None, [256], [0,256])
    return resized_image, grayscale_image, hist
```

```python
train_features = []
test_features = []
```

```python
for folder in os.listdir(train_path):
    folder_path = os.path.join(train_path, folder)
    for filename in os.listdir(folder_path):
        image_path = os.path.join(folder_path, filename)
        resized_image, grayscale_image, hist = preprocess_image(image_path)
        train_features.append((resized_image, grayscale_image, hist))
```

```python
for folder in os.listdir(test_path):
    folder_path = os.path.join(test_path, folder)
    for filename in os.listdir(folder_path):
        image_path = os.path.join(folder_path, filename)
        resized_image, grayscale_image, hist = preprocess_image(image_path)
        test_features.append((resized_image, grayscale_image, hist))
```

```python
train_image, train_gray_image, train_hist = train_features[0]
plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(train_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.subplot(1, 3, 2)
plt.imshow(train_gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.subplot(1, 3, 3)
plt.plot(train_hist)
plt.title('Histogram')
plt.show()
```

```python
print("Features of the first image in the training set:")
print("Resized Image:", train_features[0][0].shape)
print("Grayscale Image:", train_features[0][1].shape)
print("Histogram:", train_features[0][2].shape)
```

```python
test_image, test_gray_image, test_hist = test_features[0]
plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.subplot(1, 3, 2)
plt.imshow(test_gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.subplot(1, 3, 3)
plt.plot(test_hist)
plt.title('Histogram')
plt.show()
```

```python
print("\nFeatures of the first image in the testing set:")
print("Resized Image:", test_features[0][0].shape)
print("Grayscale Image:", test_features[0][1].shape)
print("Histogram:", test_features[0][2].shape)
```
