```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/CV_images/the marvels2.jpg')
```

```python
cv2_imshow(img)
```

```python
print('Image Width is',img.shape[1])
print('Image Height is',img.shape[0])
```

```python
cv2.resize(img, (500,300))
```

```python
rows,cols,_ = img.shape
```

```python
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
```

```python
M = cv2.getAffineTransform(pts1,pts2)
```

```python
dst = cv2.warpAffine(img,M,(cols,rows))
cv2_imshow(dst)
```

```python
img1 = cv2.imread('/content/drive/MyDrive/CV_images/the marvels2.jpg', 0)
```

```python
cv2_imshow(img1)
```

```python
rows,cols = img.shape
pts1 = np.float32([[150,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[10,250]])
```

```python
M1 = cv2.getAffineTransform(pts1,pts2)
dst1 = cv2.warpAffine(img1,M1,(cols,rows))
```

```python
plt.subplot(121), plt.imshow(img1, cmap='gray' ), plt.title('Input')
plt.subplot(122), plt.imshow(dst1, cmap='gray'), plt.title('Output')
plt.show()
```
