```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
image = cv2.imread('/content/drive/MyDrive/CV_images/the marvels3.jpg')
```

```python
cv2_imshow(image)
```

```python
krnlblur = np.ones((5, 5), np.float32)/30
```

```python
img = cv2.filter2D(src=image, ddepth=-1, kernel=krnlblur)
```

```python
cv2_imshow(img)
```

```python
krnlidentity = np.array([[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]])
```

```python
img1 = cv2.filter2D(src=image, ddepth=-1, kernel=krnlidentity)
```

```python
cv2_imshow(img1)
```

```python
edgekrnl = np.array([[-1, -1, -1],
                    [-1, 8, -1],
                    [-1, -1, -1]])
```

```python
img2 = cv2.filter2D(src=image, ddepth=-1, kernel=edgekrnl)
```

```python
cv2_imshow(img2)
```

```python
sharpenkrnl = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
```

```python
img3 = cv2.filter2D(src=image, ddepth=-1, kernel=sharpenkrnl)
```

```python
cv2_imshow(img3)
```

```python
boxblurkrnl = np.array([[1, 1, 1],
                    [1, 1, 1],
                    [1, 1, 1]])/9
```

```python
img4 = cv2.filter2D(src=image, ddepth=-1, kernel=boxblurkrnl)
```

```python
cv2_imshow(img4)
```

```python
gaussianblurkrnl = np.array([[1, 4, 6, 4, 1],
                             [4, 16, 24, 16, 4],
                             [6, 24, 36, 24, 6],
                             [4, 16, 24, 16, 4],
                             [1, 4, 6, 4, 1]])/256
```

```python
img5 = cv2.filter2D(src=image, ddepth=-1, kernel=gaussianblurkrnl)
```

```python
cv2_imshow(img5)
```

```python
from scipy import ndimage
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_path = '/content/drive/MyDrive/CV_images/the marvels3.jpg'
image = np.array(Image.open(image_path).convert('L'))  # Convert to grayscale
```

```python
modes = ['reflect', 'constant', 'nearest', 'mirror', 'wrap']

plt.figure(figsize=(15, 8))

for i, mode in enumerate(modes, 1):
    convolved_image = ndimage.convolve(image, krnlblur, mode=mode, cval=0.0)

    plt.subplot(2, 3, i)
    plt.imshow(convolved_image, cmap='gray')
    plt.title(f'Mode: {mode}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
