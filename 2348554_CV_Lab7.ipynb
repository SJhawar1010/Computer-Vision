```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Gaussian Noise


```python
mean = 0
var = 100
sigma = var ** 0.5
rows, cols, channels = img.shape
gauss = np.random.normal(mean, sigma, (rows, cols, channels))
noisy_img = img + gauss.astype(np.uint8)
```

```python
noisy_img
```

```python
from scipy.ndimage import minimum_filter
min=minimum_filter(img, size=3)
plt.imshow(min)
```

```python
from scipy.ndimage import maximum_filter
max=maximum_filter(img, size=3)
plt.imshow(max)
```

```python
from scipy.ndimage import median_filter
mid=median_filter(img, size=3)
plt.imshow(mid)
```
